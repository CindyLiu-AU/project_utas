```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
np.random.seed(42)
```

# load  test data

```python
df1_test_ori=pd.read_csv('BBB_planC_monthly_testset.csv', low_memory=False)
```

```python
df1_test_ori.head(13)
```

```python
df1_test_sorted=df1_test_ori.sort_values(by=['id1','act_type'])
df1_test_sorted.head(13)
```

```python
df1_test_sorted.info() # final_result  is an object
```

# make a list for storing labels

```python
def make_labels(df):
    labels=[]
    for x in range(0,12): # there are 12 activities as rows for each student id
        labels.append(list(df.iloc[0:x+1,1:2].values[x]))
    return labels
```

```python
test_labels=[]
for num in df1_test_sorted['id1'].unique():    
    t1 = make_labels(df1_test_sorted[df1_test_sorted['id1']==num]) 
    del t1[11]
    del t1[10]
    print("student",num,"label:",t1) 
    test_labels=test_labels+t1
```

```python
len(test_labels)
```

# export train_y

```python
df_test_labels = pd.DataFrame (test_labels, columns = ['final_result'])
```

```python
df_test_labels.to_csv('BBB_planD_monthly_testset_y.csv')
```

# make a list for storing student id

```python
def make_id(df):
    id=[]
    for i in range(0,12): # there are 12 activities as rows for each student id
        id.append(list(df.iloc[:,-1:].values[i]))
    return id
```

```python
test_id=[]
for num1 in df1_test_sorted['id1'].unique():    
    t2 = make_id(df1_test_sorted[df1_test_sorted['id1']==num1]) 
    del t2[11]
    del t2[10]
    print("student",num1,"label:",t2) 
    test_id=test_id+t2
```

```python
len(test_id)
```

# prepare data for transpose 

```python
df1_test_X=df1_test_sorted.drop('final_result', axis='columns')
df1_test_X=df1_test_X.drop('act_type', axis='columns')
df1_test_X
```

# transposing data

```python
def make_transpose(df):
    df_transpose=[]
    for x in range(0,12): # there are 12 activities as rows for each student id
        df_transpose=df.T
    return df_transpose
```

```python
df_test_transpose=pd.DataFrame({})
for num in df1_test_X['id1'].unique(): 
    t = make_transpose(df1_test_X[df1_test_X['id1']==num])
    print("student",num,"index:",t)
    t = t.set_axis(['Act1', 'Act2', 'Act3', 'Act4', 'Act5','Act6', 'Act7', 'Act8', 'Act9', 'Act10','Act11', 'Act12'], axis=1, inplace=False)
    df_test_transpose = df_test_transpose.append(t)
    #     df_train_transpose = df_train_transpose.append(t, ignore_index=True)


    
    
```

```python
df_test_transpose
```

```python
for x in range(10, 111):
    j=10
    df_test_transpose_sorted = df_test_transpose.drop(df_test_transpose.index[j])
    j=j+11
    
df_test_transpose_sorted=df_test_transpose_sorted.reset_index()
df_test_transpose_sorted= df_test_transpose_sorted.rename(columns={"index":"TimePeriod"})
df_test_transpose_sorted
```

# export train_X

```python
planD_test_X = df_test_transpose_sorted
planD_test_X.to_csv('BBB_planD_monthly_testset_X.csv')
```
