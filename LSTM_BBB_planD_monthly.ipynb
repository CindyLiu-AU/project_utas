# reshape X and y

```python
print(type(X_train))
print(type(y_train))
print(type(X_test))
print(type(y_test))
```

```python
X_train_reshaped=X_train.values.reshape((X_train.shape[0], X_train.shape[1] ))
X_test_reshaped=X_test.values.reshape((X_test.shape[0],X_test.shape[1]))

y_train_reshaped=y_train.values.reshape((y_train.shape[0],1 ))
y_test_reshaped=y_test.values.reshape((y_test.shape[0],1))
```

```python
print("xtrain_reshaped:", X_train_reshaped.shape)
print("xtrain_reshaped:", type(X_train_reshaped))
print("xtest_reshaped:", X_test_reshaped.shape)
print("xtest_reshaped:", type(X_test_reshaped))
```

```python
print("ytrain_reshaped:", y_train_reshaped.shape)
print("ytrain_reshaped:", type(y_train_reshaped))
print("ytest_reshaped:", y_test_reshaped.shape)
print("ytest_reshaped:", type(y_test_reshaped))
```

# scale X to be (0,1)

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_train_scaled = min_max_scaler.fit_transform(X_train_reshaped)
X_test_scaled = min_max_scaler.fit_transform(X_test_reshaped)
```

# Reshape X again 

```python
X_train_reshaped2=X_train_scaled.reshape((3737,10,12))
X_test_reshaped2=X_test_scaled.reshape((1604,10,12))
```

```python
print("xtrain_reshaped2:", X_train_reshaped2.shape)
print("xtrain_reshaped2:", type(X_train_reshaped2))
print("xtest_reshaped2:", X_test_reshaped2.shape)
print("xtest_reshaped2:", type(X_test_reshaped2))
```

# Reshape y again 

```python
y_train_reshaped2=y_train_reshaped.reshape((3737,10,1))
y_test_reshaped2=y_test_reshaped.reshape((1604,10,1))
```

```python
print("ytrain_reshaped2:", y_train_reshaped2.shape)
print("ytrain_reshaped2:", type(y_train_reshaped2))
print("ytest_reshaped2:", y_test_reshaped2.shape)
print("ytest_reshaped2:", type(y_test_reshaped2))
```

```python
y_test_squeezed = np.squeeze(y_test_reshaped2, axis=2)

arr = y_test_squeezed.reshape(-1)

y_test_new=[]
j=0
for i in range(0,y_test_squeezed.shape[0]):
    y_test_new.append(arr[j])
    j=j+10
    
len(y_test_new)
```

```python
arr_y_test_new = np.array(y_test_new)
arr_y_test_new=arr_y_test_new.reshape(1604,1)
arr_y_test_new.shape
```

```python
y_train_squeezed = np.squeeze(y_train_reshaped2, axis=2)

arr = y_train_squeezed.reshape(-1)

y_train_new=[]
j=0
for i in range(0,y_train_squeezed.shape[0]):
    y_train_new.append(arr[j])
    j=j+10
    
len(y_train_new)
```

```python
arr_y_train_new = np.array(y_train_new)
arr_y_train_new=arr_y_train_new.reshape(3737,1)
arr_y_train_new.shape
```

# the final X and y for modelling

```python
X_train1 = X_train_reshaped2
X_test1 = X_test_reshaped2
y_train1 = arr_y_train_new
y_test1 = arr_y_test_new
```

```python
print(X_train1.shape)
print(X_test1.shape)
print(y_train1.shape)
print(y_test1.shape)
```

# -----------------The second file: LSTM-------------------

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM, Masking, Dropout
from keras.preprocessing import sequence
import tensorflow as tf
from tensorflow import keras
np.random.seed(42)
tf.random.set_seed(42)
```

```python
batch_size = 128
```

```python
## 
    
lr=0.0001
epochs = 100

model = Sequential()
model.add(LSTM(40, dropout=0.2, input_shape=(X_train1.shape[1], X_train1.shape[2]),return_sequences=True))
model.add(LSTM(20, dropout=0.2, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
              metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
history_model = model.fit(X_train1, y_train1, validation_data=(X_test1, y_test1), epochs=epochs, batch_size=batch_size)
```

```python
history_model.params
```

```python
model.evaluate(X_test1, y_test1)
```

# plot learning process

```python
import matplotlib.pyplot as plt

pd.DataFrame(history_model.history).plot(figsize=(10, 7))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.show()
```

```python
import matplotlib.pyplot as plt
def validation_training_process(history):
  plt.plot(history.history['loss'], label='Training Loss', color='green', linestyle = '--')
  plt.plot(history.history['val_loss'], label='Validation Loss', color='blue', linestyle ='--')
  plt.plot(history.history['accuracy'], label='Training Accuracy', color='green')
  plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='blue')
  plt.title('Training Process Visualization')

  plt.xlabel('No. epoch')
  plt.legend()
  plt.show()
```

```python
validation_training_process(history_model)
```
